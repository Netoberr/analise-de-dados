```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sales_data = pd.read_csv("https://raw.githubusercontent.com/Jeferson-Peter/MLs05a-Celsius-para-Fahrenheit/master/MLs05a%20Celsius%20para%20Fahrenheit.csv") 
sales_data
```

Etapa 4: Configuração da base de dados de treinamento

```python
x_train = sales_data['Celsius']
y_train = sales_data['Fahrenheit']
```

```python
#Mostra basicamente o tamanho dos dados, ou seja, tem 30 dados, sao 30 especificos
x_train.shape
```

Etapa 5: construção do modelo

```python
#Feature => coluna Celsius ou Fahrenheit
# Param  
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=10, input_shape=[1]))
model.add(tf.keras.layers.Dense(units=5, input_shape=[1]))
model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(0.1), loss='mean_squared_error')
epochs_hist = model.fit(x_train, y_train, epochs=100)
```
