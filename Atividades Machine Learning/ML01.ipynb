```python
dados
```

```python
from sklearn.svm import LinearSVC

model = LinearSVC()
model.fit(dados, classes)
```

```python
animal_misterioso = [1,1,1]
model.predict([animal_misterioso])
```

```python
misterio1 = [1,1,1]
misterio2 = [1,1,0]
misterio3 = [0,1,1]

testes = [misterio1, misterio2, misterio3]
previsoes = model.predict(testes)
```

```python
testes_classes = [0,1,1]
```

```python
corretos = (previsoes == testes_classes).sum()
total = len(testes)
taxa_de_Acerto = corretos/total
print(f'Taxa de acerto {taxa_de_Acerto * 100}')
```

```python
from sklearn.metrics import accuracy_score

taxa_de_acerto = accuracy_score(testes_classes, previsoes)
print(f'Taxa de acerto: {taxa_de_acerto *100}')
```

**Usando biblioteca para separar treino e teste**



*   Randon com a definição de um seed faz sempre utilizarmos o mesmo conjunto pego de forma aleatória
*   Para não alterar o resultado a cada execução - para que o experimento seja replicável



```python
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

SEED = 11

treino_x, teste_x, treino_y, teste_y = train_test_split(x,y, random_state=SEED, 
                                                        test_size = 0.25)
print(f"""Treinaremos com {len(treino_x)} elementos e testaremos com
{len(treino_x)}""")

modelo = LinearSVC()
modelo.fit(treino_x, treino_y)
print(treino_x)
previsoes = modelo.predict(teste_x)
taxa_de_acerto = accuracy_score(teste_y, previsoes) * 100
print(f'Taxa de acerto: {round(taxa_de_acerto,2)}%')
```
