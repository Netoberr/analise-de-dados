```python
%tensorflow_version 2.x
import tensorflow as tf
tf.__version__
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sales_data = pd.read_csv("https://raw.githubusercontent.com/Jeferson-Peter/MLs05b-SalesData/master/MLs05b%20SalesData.csv") 
sales_data
```

```python
sales_data.describe()
```

```python
sns.scatterplot(x=sales_data['Temperature'], y=sales_data[ 'Revenue'])
```

```python
x_train = sales_data['Temperature']
y_train = sales_data['Revenue']
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
#Feature => coluna Celsius ou Fahrenheit
# Param  
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=10, input_shape=[1]))
model.add(tf.keras.layers.Dense(units=5, input_shape=[1]))
model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(0.1), loss='mean_squared_error')
epochs_hist = model.fit(x_train, y_train, epochs=100)
```

```python
epochs_hist.history
```

```python
plt.plot(epochs_hist.history['loss'])
plt.title('Model Loss Progress During Training')
plt.xlabel('Epoch')
plt.ylabel('Training loss')
plt.legend(['Training loss'])
```

```python
model.get_weights()
```

```python
temp_c = 25
predict = model.predict([temp_c])
predict
```
