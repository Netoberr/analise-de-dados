Converter variáveis categóricas em numéricas
1) Usar LabelEncoder() para converter os dados em números:

```python
label_encoder = LabelEncoder()
labels_zoning = label_encoder.fit_transform(df.MSZoning)
labels_zoning
```

```python
df['Zoneamento'] = labels_zoning
df.head(10)
```

## Criar colunas binárias ara cada valor categórico

```python
one_hot_encoder = OneHotEncoder()
feature_array = one_hot_encoder.fit_transform(df[['Zoneamento']]).toarray()
feature_array
```

## "Juntar" o array com o Dataset

```python
feature_labels = list(label_encoder.classes_)
features = pd.DataFrame(feature_array, columns=feature_labels)
features
```

```python
df = pd.concat([df, features], axis=1)
df
```

Fazer para coluna Street

```python
label_encoder = LabelEncoder()
labels_streeting = label_encoder.fit_transform(df.Street)
labels_streeting
```

```python
df['Rua'] = labels_streeting
df.head(10)
```

```python
one_hot_encoder = OneHotEncoder()
feature_array = one_hot_encoder.fit_transform(df[['Rua']]).toarray()
feature_array
```

```python
feature_labels = list(label_encoder.classes_)
features = pd.DataFrame(feature_array, columns=feature_labels)
features
```

```python
df = pd.concat([df, features], axis=1)
df
```
